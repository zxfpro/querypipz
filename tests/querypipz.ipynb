```python
from querypipz.factory.store import GraphStore,GraphStoreType
```

```python
gs = GraphStore(GraphStoreType.Neo4jGraphStore)
```

```python
# 上传一个节点
```

```python
from llama_index.core.graph_stores.types import EntityNode, ChunkNode, Relation

# Create a two entity nodes
entity1 = EntityNode(name="工人",      properties={"age": 32}, label="PERSON",  )
entity2 = EntityNode(name="works",  properties={"age": 32, "pgtext": [3,4,5,6]},  label="ORGANIZATION", )

# Create a relation
relation = Relation(
    label="WORKS_FOR",
    source_id=entity1.id,
    target_id=entity2.id,
    properties={"since": 2023},
)
```

```python
gs.upsert_nodes([entity1, entity2])
gs.upsert_relations([relation])
```
